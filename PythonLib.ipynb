**Shapping **

```python
import numpy as np
arr=np.arange(6)
reshaped=arr.reshape((2,3))
print(reshaped)
reshaped=arr.reshape((3,2))
print(reshaped)
```

```python
arr=np.array([[1,2],[3,4]])
flat=arr.flatten()
print(flat)

arr=np.array([[1,2],[3,4]])
flat=arr.flatten()
print(flat)
```

```python
arr1 = np.array([[1, 2], [3, 4]])
arr2 = np.array([[5, 6], [7, 8]])

merged = np.concatenate([arr1, arr2], axis=0)
print(merged)

merged = np.concatenate([arr1, arr2], axis=1)
print(merged)
# axis =0 row-wise (vertical)
# axis=1 column-wise (horizontal)
```

```python
arr1 = np.array([[1, 2,3], [4,5,6]])
arr2 = arr1.T
print(arr2)
```

```python
arr=np.array([0,2,0,5,6,0,9])
count=np.count_nonzero(arr)
print(count)
```

```python
arr=np.array([1,np.nan,2,np.nan,3,4,5])
cleaned=np.nan_to_num(arr,nan=0)
print(cleaned)
```

```python
arr=np.array([1,np.nan,2,np.nan,3,4,5])
nan_mask=np.isnan(arr)
print(nan_mask)
```
